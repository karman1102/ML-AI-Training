```python
import pandas as pd
```

```python
data = pd.read_csv('xclara.csv')
```

```python
data.head()
```

```python
import numpy as np
```

```python
c1 = data['V1'].values
```

```python
c2 = data['V2'].values
```

```python
data.shape
```

```python
from sklearn.cluster import KMeans
```

```python
range_cluster = [1,2,3,5,7,10,15] #define a rage for the number of clusters to find the suitable K
```

```python
c1
```

```python
sse = {} #making a dectionary like a hash map to  store cost
```

```python
for n_clusters in range_cluster:
    kmeans = KMeans(n_clusters = n_clusters) #kmeans is assigned a value of a the range assigned to number of clusters in a function of KMean
    kmeans_label = kmeans.fit(data) #each data when fit is then given a label to feed ito the dictionary
    sse[n_clusters] = kmeans_label.inertia_ #works like score which feeds the score in the ddictionary
```

```python
#matplotlib inline
from matplotlib import pyplot as plt
plt.figure()
plt.plot(list(sse.keys()), list(sse.values()))
plt.xlabel('Number of cluster')
plt.ylabel('COST')
plt.show()
```

```python
plt.scatter(c1,c2 ,color = 'black')
```
